```python
import pandas as pd

# ── Step 1: Keep the core case-level columns ──────────────────────
case_cols = [
    'Case ID', 'Resource', 'Complete Timestamp', 'Variant', 'Variant index',
    'OrganizationalEntity', 'travel permit number', 'RequestedBudget',
    'BudgetNumber', 'TotalDeclared', 'OverspentAmount', 'Overspent',
    'org:role', 'id', 'id.1'
]
# Only keep ones that actually exist
case_cols = [c for c in case_cols if c in travel_permits.columns]

# ── Step 2: Find all numbered activity columns ────────────────────
activity_cols = [c for c in travel_permits.columns if c.startswith('Activity_')]
amount_cols   = [c for c in travel_permits.columns if c.startswith('RequestedAmount_')]
dec_id_cols   = [c for c in travel_permits.columns if c.startswith('dec_id_')]
decnum_cols   = [c for c in travel_permits.columns if c.startswith('DeclarationNumber_')]

print(f"Activity columns found:   {len(activity_cols)}")
print(f"Amount columns found:     {len(amount_cols)}")
print(f"Dec ID columns found:     {len(dec_id_cols)}")
print(f"Declaration num columns:  {len(decnum_cols)}")

# ── Step 3: Melt into long format ─────────────────────────────────
# Extract the index number from each column name (Activity_0 → 0)
records = []

for _, row in travel_permits.iterrows():
    case_data = {c: row[c] for c in case_cols}

    # Loop over each activity slot
    for i in range(len(activity_cols)):
        activity_val = row.get(f'Activity_{i}', None)

        # Skip empty slots
        if pd.isna(activity_val) or activity_val == '':
            continue

        event = case_data.copy()
        event['event_index']        = i
        event['Activity']           = activity_val
        event['RequestedAmount']    = row.get(f'RequestedAmount_{i}', None)
        event['dec_id']             = row.get(f'dec_id_{i}', None)
        event['DeclarationNumber']  = row.get(f'DeclarationNumber_{i}', None)
        records.append(event)

permits_long = pd.DataFrame(records)

print(f"\nOriginal shape:  {travel_permits.shape}")
print(f"Long format shape: {permits_long.shape}")
print(f"\nSample:")
print(permits_long[['Case ID', 'travel permit number', 'Activity', 'RequestedAmount', 'Complete Timestamp']].head(10).to_string(index=False))
```

```python
# ── Step 1: Convert timestamps in all 5 tables ───────────────────

# Travel Permits — already renamed to 'timestamp' in reshape step
permits_eventlog['timestamp'] = pd.to_datetime(
    permits_eventlog['timestamp'], errors='coerce'
)
print(f"  permits_eventlog            dtype: {permits_eventlog['timestamp'].dtype}  "
      f"parse failures: {permits_eventlog['timestamp'].isnull().sum()}")

# Remaining 4 tables — still use original column name
other_tables = {
    'intl_expense_claims':      intl_expense_claims,
    'national_expense_claims':  national_expense_claims,
    'travel_expenses':          travel_expenses,
    'payment_requests':         payment_requests,
}

for name, df in other_tables.items():
    df['Complete Timestamp'] = pd.to_datetime(
        df['Complete Timestamp'], errors='coerce'
    )
    nulls = df['Complete Timestamp'].isnull().sum()
    print(f"  {name:<30} dtype: {df['Complete Timestamp'].dtype}  "
          f"parse failures: {nulls}")

print("\nAll timestamps converted successfully.")
```

```python
# ── Step 2a: Clean Travel Permits event log ───────────────────────
# Already renamed in reshape step — just select the columns we need
permits_clean = permits_eventlog[[
    'case_id',
    'activity',
    'timestamp',
    'resource',
    'role',
    'variant',
    'variant_index',
    'permit_number',
    'requested_budget',
    'total_declared',
    'overspent_amount',
    'overspent',
]].copy()

print(f"Permits clean shape:    {permits_clean.shape}")
print(f"Columns: {permits_clean.columns.tolist()}")
```

```python
# ── Step 2b: Clean International Expense Claims ───────────────────
intl_clean = intl_expense_claims[[
    'Case ID',
    'Activity',
    'Complete Timestamp',
    'Resource',
    'org:role',
    'Variant',
    'Variant index',
    'Permit travel permit number',
    'DeclarationNumber',
    'Amount',
    'AdjustedAmount',
    'Permit RequestedBudget',
    'Permit OrganizationalEntity',
    'Permit BudgetNumber',
]].copy()

intl_clean.columns = [
    'case_id', 'activity', 'timestamp', 'resource', 'role',
    'variant', 'variant_index', 'permit_number', 'declaration_number',
    'amount', 'adjusted_amount', 'permit_budget',
    'org_unit', 'budget_number',
]

print(f"Intl clean shape:       {intl_clean.shape}")
print(f"Columns: {intl_clean.columns.tolist()}")
```

```python
# ── Step 2c: Clean National Expense Claims ────────────────────────
national_clean = national_expense_claims[[
    'Case ID',
    'Activity',
    'Complete Timestamp',
    'Resource',
    'org:role',
    'Variant',
    'Variant index',
    'DeclarationNumber',
    'Amount',
]].copy()

national_clean.columns = [
    'case_id', 'activity', 'timestamp', 'resource', 'role',
    'variant', 'variant_index', 'declaration_number', 'amount',
]

print(f"National clean shape:   {national_clean.shape}")
print(f"Columns: {national_clean.columns.tolist()}")
```

```python
# ── Step 2d: Clean Travel Expenses ────────────────────────────────
travel_exp_clean = travel_expenses[[
    'Case ID',
    'Activity',
    'Complete Timestamp',
    'Resource',
    'org:role',
    'Variant',
    'Variant index',
    'Permit travel permit number',
    'RfpNumber',
    'RequestedAmount',
    'Permit RequestedBudget',
    'OrganizationalEntity',
    'Permit OrganizationalEntity',
]].copy()

travel_exp_clean.columns = [
    'case_id', 'activity', 'timestamp', 'resource', 'role',
    'variant', 'variant_index', 'permit_number', 'rfp_number',
    'amount', 'permit_budget', 'org_unit', 'permit_org_unit',
]

print(f"Travel exp clean shape: {travel_exp_clean.shape}")
print(f"Columns: {travel_exp_clean.columns.tolist()}")
```

```python
# ── Step 2e: Clean Payment Requests ──────────────────────────────
payment_clean = payment_requests[[
    'Case ID',
    'Activity',
    'Complete Timestamp',
    'Resource',
    'org:role',
    'Variant',
    'Variant index',
    'RfpNumber',
    'RequestedAmount',
    'OrganizationalEntity',
    'Project',
]].copy()

payment_clean.columns = [
    'case_id', 'activity', 'timestamp', 'resource', 'role',
    'variant', 'variant_index', 'rfp_number', 'amount',
    'org_unit', 'project',
]

print(f"Payment clean shape:    {payment_clean.shape}")
print(f"Columns: {payment_clean.columns.tolist()}")
```

```python
# ── Step 3: Build unified event log ──────────────────────────────
permits_clean['process_type']    = 'travel_permit'
intl_clean['process_type']       = 'intl_expense_claim'
national_clean['process_type']   = 'national_expense_claim'
travel_exp_clean['process_type'] = 'travel_expense'
payment_clean['process_type']    = 'payment_request'

common_cols = [
    'case_id', 'activity', 'timestamp', 'resource',
    'role', 'variant', 'variant_index', 'process_type', 'amount'
]

unified_log = pd.concat([
    permits_clean.reindex(columns=common_cols),
    intl_clean.reindex(columns=common_cols),
    national_clean.reindex(columns=common_cols),
    travel_exp_clean.reindex(columns=common_cols),
    payment_clean.reindex(columns=common_cols),
], ignore_index=True).sort_values(['case_id', 'timestamp'])

print(f"Unified event log shape: {unified_log.shape}")
print(f"\nCases per process type:")
print(unified_log.groupby('process_type')['case_id'].nunique().to_string())
print(f"\nDate range: {unified_log['timestamp'].min()} → {unified_log['timestamp'].max()}")
```

```python
# ── Step 4a: Build permit-level lookup table ──────────────────────
def get_first_timestamp(group, activity_name):
    match = group[group['activity'] == activity_name]['timestamp']
    return match.min() if not match.empty else pd.NaT

permit_lookup = permits_clean.groupby('case_id').apply(
    lambda g: pd.Series({
        'permit_number':    g['permit_number'].iloc[0],
        'variant':          g['variant'].iloc[0],
        'requested_budget': g['requested_budget'].iloc[0],
        'submitted':        get_first_timestamp(g, 'Permit SUBMITTED by EMPLOYEE'),
        'final_approved':   min(filter(pd.notna, [
                                get_first_timestamp(g, 'Permit FINAL_APPROVED by SUPERVISOR'),
                                get_first_timestamp(g, 'Permit FINAL_APPROVED by DIRECTOR'),
                            ]), default=pd.NaT),
        'start_trip':       get_first_timestamp(g, 'Start trip'),
        'end_trip':         get_first_timestamp(g, 'End trip'),
        'had_rejection':    g['activity'].str.contains('REJECTED').any(),
        'rejection_count':  g['activity'].str.contains('REJECTED').sum(),
        'had_director':     (g['activity'] == 'Permit FINAL_APPROVED by DIRECTOR').any(),
        'had_pre_approver': (g['activity'] == 'Permit APPROVED by PRE_APPROVER').any(),
    })
).reset_index()

print(f"Permit lookup shape:     {permit_lookup.shape}")
print(f"Null final_approved:     {permit_lookup['final_approved'].isnull().sum()}")
print(f"Null start_trip:         {permit_lookup['start_trip'].isnull().sum()}")
print(f"Null end_trip:           {permit_lookup['end_trip'].isnull().sum()}")
print(permit_lookup.head(3).to_string(index=False))
```

```python
# ── Step 4b: Build intl claims case-level table ───────────────────
intl_cases = intl_clean.groupby('case_id').apply(
    lambda g: pd.Series({
        'permit_number':      g['permit_number'].iloc[0],
        'declaration_number': g['declaration_number'].iloc[0],
        'amount':             g['amount'].iloc[0],
        'adjusted_amount':    g['adjusted_amount'].iloc[0],
        'org_unit':           g['org_unit'].iloc[0],
        'variant':            g['variant'].iloc[0],
        'decl_submitted':     get_first_timestamp(g, 'Declaration SUBMITTED by EMPLOYEE'),
        'payment_handled':    get_first_timestamp(g, 'Payment Handled'),
        'had_rejection':      g['activity'].str.contains('REJECTED').any(),
        'rejection_count':    g['activity'].str.contains('REJECTED').sum(),
    })
).reset_index()

# Flag UNKNOWN permits as unlinked
intl_cases['permit_linked'] = (
    intl_cases['permit_number'] != 'UNKNOWN'
) & intl_cases['permit_number'].notna()

# Join to permit lookup
compliance_df = intl_cases.merge(
    permit_lookup[[
        'permit_number', 'submitted', 'final_approved',
        'start_trip', 'end_trip', 'requested_budget'
    ]],
    on='permit_number',
    how='left'
)

# Derive compliance flags
compliance_df['approved_before_trip'] = (
    compliance_df['final_approved'] < compliance_df['start_trip']
)
compliance_df['days_to_declare'] = (
    compliance_df['decl_submitted'] - compliance_df['end_trip']
).dt.days
compliance_df['throughput_days'] = (
    compliance_df['payment_handled'] - compliance_df['decl_submitted']
).dt.days

print(f"Compliance table shape:  {compliance_df.shape}")
print(f"Permit linked:           {compliance_df['permit_linked'].sum()}")
print(f"Permit not linked:       {(~compliance_df['permit_linked']).sum()}")
print(f"Approved before trip:    {compliance_df['approved_before_trip'].sum()}")
print(f"Null approved_before:    {compliance_df['approved_before_trip'].isnull().sum()}")
```

```python
# ── Step 5: Feature engineering per case ─────────────────────────
def build_case_features(df, process_type):
    has_org    = 'org_unit' in df.columns
    has_amount = 'amount'   in df.columns

    return df.groupby('case_id').apply(lambda g: pd.Series({
        'process_type':      process_type,
        'variant':           g['variant'].iloc[0],
        'org_unit':          g['org_unit'].iloc[0]  if has_org    else None,
        'amount':            g['amount'].iloc[0]    if has_amount else None,
        'total_events':      len(g),
        'throughput_days':   (g['timestamp'].max() - g['timestamp'].min()).days,
        'rejection_count':   g['activity'].str.contains('REJECTED').sum(),
        'had_rejection':     g['activity'].str.contains('REJECTED').any(),
        'had_missing_role':  (g['role'] == 'MISSING').any(),
        'had_director':      g['activity'].str.contains('DIRECTOR').any(),
        'had_pre_approver':  g['activity'].str.contains('PRE_APPROVER').any(),
        'reached_payment':   g['activity'].str.contains('Payment Handled').any(),
        'year':              g['timestamp'].min().year,
    })).reset_index()

features_permits  = build_case_features(permits_clean,    'travel_permit')
features_intl     = build_case_features(intl_clean,       'intl_expense_claim')
features_national = build_case_features(national_clean,   'national_expense_claim')
features_travel   = build_case_features(travel_exp_clean, 'travel_expense')
features_payment  = build_case_features(payment_clean,    'payment_request')

all_case_features = pd.concat([
    features_permits,
    features_intl,
    features_national,
    features_travel,
    features_payment,
], ignore_index=True)

print(f"Case features shape:     {all_case_features.shape}")
print(f"\nCases per process type:")
print(all_case_features.groupby('process_type').size().to_string())
print(f"\nThroughput days summary:")
print(all_case_features.groupby('process_type')['throughput_days'].describe().to_string())
```

```python
# ── Step 6: Multi-claim permit analysis ──────────────────────────
claims_per_permit = intl_cases[
    intl_cases['permit_linked']
].groupby('permit_number').agg(
    claim_count      = ('case_id',        'count'),
    total_amount     = ('amount',         'sum'),
    first_submission = ('decl_submitted', 'min'),
    last_payment     = ('payment_handled','max'),
).reset_index()

claims_per_permit['total_cycle_days'] = (
    claims_per_permit['last_payment'] -
    claims_per_permit['first_submission']
).dt.days

multi_claim  = claims_per_permit[claims_per_permit['claim_count'] > 1]
single_claim = claims_per_permit[claims_per_permit['claim_count'] == 1]

print(f"Total permits with claims:   {len(claims_per_permit)}")
print(f"Single claim permits:        {len(single_claim)} ({round(len(single_claim)/len(claims_per_permit)*100,1)}%)")
print(f"Multi claim permits:         {len(multi_claim)} ({round(len(multi_claim)/len(claims_per_permit)*100,1)}%)")
print(f"\nMulti-claim permit stats:")
print(f"  Max claims per permit:     {multi_claim['claim_count'].max()}")
print(f"  Mean claims per permit:    {multi_claim['claim_count'].mean():.1f}")
print(f"  Mean total cycle days:     {multi_claim['total_cycle_days'].mean():.1f}")
print(f"  Single claim mean days:    {single_claim['total_cycle_days'].mean():.1f}")
```

```python
# ── Step 7: Flag anomaly candidates ──────────────────────────────
import numpy as np
from scipy import stats

anomaly_flags = []

# Flag 1 — Amount outliers using IQR per process type
for name, df in [
    ('intl_expense_claim', intl_cases),
    ('travel_expense',     features_travel),
    ('payment_request',    features_payment),
]:
    amounts  = df['amount'].dropna()
    Q1, Q3   = amounts.quantile(0.25), amounts.quantile(0.75)
    IQR      = Q3 - Q1
    upper    = Q3 + 1.5 * IQR
    lower    = Q1 - 1.5 * IQR
    outliers = df[
        (df['amount'] > upper) | (df['amount'] < lower)
    ][['case_id', 'amount']].copy()
    outliers['flag']         = 'amount_outlier_IQR'
    outliers['process_type'] = name
    anomaly_flags.append(outliers)
    print(f"  {name:<30} IQR outliers: {len(outliers)}")

# Flag 2 — MISSING role cases
missing_role_cases = unified_log[
    unified_log['role'] == 'MISSING'
][['case_id', 'process_type']].drop_duplicates().copy()
missing_role_cases['flag']   = 'missing_role'
missing_role_cases['amount'] = None
anomaly_flags.append(missing_role_cases)
print(f"  missing role cases:              {len(missing_role_cases)}")

# Flag 3 — Cases that never reached payment
no_payment = all_case_features[
    ~all_case_features['reached_payment']
][['case_id', 'process_type']].copy()
no_payment['flag']   = 'no_payment_reached'
no_payment['amount'] = None
anomaly_flags.append(no_payment)
print(f"  cases without payment:           {len(no_payment)}")

# Flag 4 — Extreme throughput time Z-score > 3
tp       = all_case_features['throughput_days'].dropna()
z_scores = np.abs(stats.zscore(tp))
extreme_tp = all_case_features.loc[
    all_case_features.index.isin(tp.index[z_scores > 3]),
    ['case_id', 'process_type', 'throughput_days']
].copy()
extreme_tp['flag']   = 'extreme_throughput_zscore'
extreme_tp['amount'] = None
anomaly_flags.append(extreme_tp)
print(f"  extreme throughput cases:        {len(extreme_tp)}")

# Combine all flags
anomaly_candidates = pd.concat(anomaly_flags, ignore_index=True)
print(f"\nTotal anomaly candidate rows:    {len(anomaly_candidates)}")
print(f"Unique flagged cases:             {anomaly_candidates['case_id'].nunique()}")
print(f"\nFlags breakdown:")
print(anomaly_candidates['flag'].value_counts().to_string())
```
